```python
import pandas as pd

# Load the dataset
df = pd.read_csv("D:/football analytics/very latest single large dataset/latest Trimmed Path/Final_trimmed_Playerstats_from_2015to2024.csv")

# Define the columns to keep for each category

attacking_columns_to_keep = [
    "Year", "ID", "Name", 
    "Finishing", "ShotPower", "Positioning", "Crossing",
    "Dribbling", "BallControl", "Vision", "LongShots"
]
midfield_columns_to_keep = [
    "Year", "ID", "Name",  
    "ShortPassing", "Vision", "BallControl", 
    "LongPassing", "Stamina", "Reactions", "Composure","Dribbling"
]
defending_columns_to_keep = [
    "Year", "ID", "Name",   
    "Marking", "StandingTackle", "SlidingTackle", "Interceptions",
    "Strength", "Aggression", "Jumping", "Composure"
]
goalkeeping_columns_to_keep = [
    "Year", "ID", "Name",  
    "GKDiving", "GKHandling", "GKKicking", 
    "GKPositioning", "GKReflexes", "Jumping"
]
economical_columns_to_keep = ["ID", "Year", "Name", "Value", "Wage","Release Clause"]

playerstats_columns_to_keep = [
    "Year", "ID", "Name", "Nationality", "Age", "Height", "Weight", 
    "Preferred Foot", "Body Type", "Skill Moves", "Weak Foot", 
    "Finishing", "ShotPower", "Positioning", "Crossing", "Marking", 
    "StandingTackle", "SlidingTackle", "Interceptions", "ShortPassing", 
    "Vision", "BallControl", "LongPassing", "Stamina", "GKDiving", 
    "GKHandling", "GKKicking", "GKPositioning", "GKReflexes", "Value", 
    "Wage", "Contract Valid Until"
]

physicalfitness_columns_mapping = [

 "Year", "ID", "Name","Height", "Weight", "Strength","Stamina","Agility", "Composure"
]


# Filter the data
attacking_df = df[attacking_columns_to_keep]
midfield_df = df[midfield_columns_to_keep]
defending_df = df[defending_columns_to_keep]
goalkeeping_df = df[goalkeeping_columns_to_keep]
economical_stats_df = df[economical_columns_to_keep]
playerstats_2024_df=df[playerstats_columns_to_keep]
physicalfit_stats_df=df[physicalfitness_columns_mapping]

# Calculate Attacking Score
attackscoreformula= (
    (0.2 * attacking_df['Finishing']) + 
    (0.2 * attacking_df['ShotPower']) + 
    (0.15 * attacking_df['Dribbling']) + 
    (0.15 * attacking_df['BallControl']) + 
    (0.1 * attacking_df['Crossing']) + 
    (0.1 * attacking_df['Vision']) + 
    (0.1 * attacking_df['LongShots']) + 
    (0.1 * attacking_df['Positioning'])
)

# Calculate Defending Score
defendscoreformula=(
    (0.25 * defending_df['Interceptions']) + 
    (0.2 * defending_df['Marking']) + 
    (0.2 * defending_df['StandingTackle']) + 
    (0.1 * defending_df['SlidingTackle']) + 
    (0.1 * defending_df['Strength']) + 
    (0.05 * defending_df['Aggression']) + 
    (0.05 * defending_df['Jumping']) + 
    (0.05 * defending_df['Composure'])
)

# Calculate Fielding Score
fieldscoreformula=(
    (0.2 * midfield_df['ShortPassing']) + 
    (0.15 * midfield_df['LongPassing']) + 
    (0.15 * midfield_df['Vision']) + 
    (0.15 * midfield_df['BallControl']) + 
    (0.1 * midfield_df['Dribbling']) + 
    (0.1 * midfield_df['Stamina']) + 
    (0.1 * midfield_df['Reactions']) + 
    (0.05 * midfield_df['Composure'])
)

# Calculate Goalkeeping Score
keepscoreformula=(
    (0.25 * goalkeeping_df['GKDiving']) + 
    (0.2 * goalkeeping_df['GKHandling']) + 
    (0.2 * goalkeeping_df['GKReflexes']) + 
    (0.15 * goalkeeping_df['GKPositioning']) + 
    (0.1 * goalkeeping_df['GKKicking']) + 
    (0.1 * goalkeeping_df['Jumping'])
)



attacking_df['AttackingScore'] = attackscoreformula
defending_df['DefendingScore'] = defendscoreformula
midfield_df['FieldingScore'] = fieldscoreformula
goalkeeping_df['GoalkeepingScore'] = keepscoreformula

#playerstats_20224 columns appending
playerstats_2024_df['AttackingScore'] = attackscoreformula
playerstats_2024_df['DefendingScore'] = defendscoreformula
playerstats_2024_df['FieldingScore'] = fieldscoreformula
playerstats_2024_df['GoalkeepingScore'] = keepscoreformula


attacking_file_path = "D:/football analytics/very latest single large dataset/latest Trimmed Path/data to be used by powerbi/recent replication/attacking abilities.csv"
attacking_df.to_csv(attacking_file_path, index=False)
print(f"Attacking abilities saved to {attacking_file_path}")

# Save each DataFrame
attacking_file_path = "D:/football analytics/very latest single large dataset/latest Trimmed Path/data to be used by powerbi/recent replication/attacking abilities.csv"
attacking_df.to_csv(attacking_file_path, index=False)
print(f"Attacking abilities saved to {attacking_file_path}")

midfield_file_path = "D:/football analytics/very latest single large dataset/latest Trimmed Path/data to be used by powerbi/recent replication/fielding abilities.csv"
midfield_df.to_csv(midfield_file_path, index=False)
print(f"Midfield abilities saved to {midfield_file_path}")

defending_file_path = "D:/football analytics/very latest single large dataset/latest Trimmed Path/data to be used by powerbi/recent replication/defending abilities.csv"
defending_df.to_csv(defending_file_path, index=False)
print(f"Defending abilities saved to {defending_file_path}")

goalkeeping_file_path = "D:/football analytics/very latest single large dataset/latest Trimmed Path/data to be used by powerbi/recent replication/goalkeeping abilities.csv"
goalkeeping_df.to_csv(goalkeeping_file_path, index=False)
print(f"Goalkeeping abilities saved to {goalkeeping_file_path}")

economical_file_path = "D:/football analytics/very latest single large dataset/latest Trimmed Path/data to be used by powerbi/recent replication/economical stats.csv"
economical_stats_df.to_csv(economical_file_path, index=False)
print(f"Economical stats saved to {economical_file_path}")


playerstats_2024_df = playerstats_2024_df[playerstats_2024_df['Year'] == 24]
playerstats_2024_file_path = "D:/football analytics/very latest single large dataset/latest Trimmed Path/data to be used by powerbi/recent replication/playerstats_2024.csv"
playerstats_2024_df.to_csv(playerstats_2024_file_path, index=False)
print(f"playerstats_2024 stats saved to {playerstats_2024_file_path}")

physicalfit_stats_df = physicalfit_stats_df[physicalfit_stats_df['Year'] == 24]
physicalfit_stats_file_path = "D:/football analytics/very latest single large dataset/latest Trimmed Path/data to be used by powerbi/recent replication/playerfitness.csv"
physicalfit_stats_df.to_csv(physicalfit_stats_file_path, index=False)
print(f"playerstats_2024 stats saved to {physicalfit_stats_file_path}")


```
